 4. Desenvolver um programa que crie um array de objetos - Automovel com três tipos do mesmo.
   * Exemplo:
   * Objeto: Automovel
   * carros[0] = ("Fiat", "Palio", "Branco");  
   * carros[1] = ("Ford", "KA",    "Branco");
   * carros[2] = ("VW",   "Jetta", "Branco");

```python
# 06/10/2023
# Brennon Meireles
# Senai Suiço Brasileiro
# Exercício de FPOO

carros = 0 # var para armazenar os valores dos carros

quantidade = int(input('Digite o número de Carros que serão catalogados: '))

catalogo = []

for i in range(quantidade):
  carros += 1

  aux = []

  # var para definir a fabricante do carro:
  fabricante = str(input('\nDigite a fabricante do carro: '))
  aux.append(fabricante)

  # var para definir o modelo do carro:
  modelo = str(input('Digite o modelo do carro: '))
  aux.append(modelo)

  # var para definir a cor do carro:
  cor = str(input('Digite a cor do carro: '))
  aux.append(cor)

  catalogo.append(aux)

print(f'\nCarros que serão cadastrados: {carros}')
print('\nItens cadastrados no Catálogo: ')
for carro in catalogo:
  print(f'Fabricante: {carro[0]}, Modelo: {carro[1]} ,Cor: {carro[2]}')
```
